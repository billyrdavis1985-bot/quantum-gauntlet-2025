```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns

# --- AUTOMATED NOVELTY SCORING (No manual judge needed) ---
def calculate_novelty_scores(question_idx):
    """
    Novelty = How different from other responses + unique vocabulary richness
    """
    responses = [data["responses"][model][question_idx] for model in model_names]

    # TF-IDF to find unique phrases
    vectorizer = TfidfVectorizer(ngram_range=(1,3), max_features=100)
    tfidf_matrix = vectorizer.fit_transform(responses)

    # Cosine similarity (lower = more novel)
    similarities = cosine_similarity(tfidf_matrix)

    # Novelty score = inverse of average similarity to others
    novelty_scores = {}
    for i, model in enumerate(model_names):
        # Average similarity to OTHER responses (exclude self)
        other_sims = [similarities[i][j] for j in range(4) if j != i]
        avg_similarity = np.mean(other_sims)
        novelty_scores[model] = 1 - avg_similarity  # Higher = more novel

    # Vocabulary richness (unique words / total words)
    for i, model in enumerate(model_names):
        words = responses[i].lower().split()
        unique_ratio = len(set(words)) / len(words) if words else 0
        novelty_scores[model] *= (1 + unique_ratio)  # Boost for rich vocabulary

    return novelty_scores

# --- MULTI-METRIC ANALYSIS ---
def comprehensive_novelty_analysis():
    """
    Analyzes novelty across ALL questions, creates heatmap + rankings
    """
    novelty_matrix = np.zeros((7, 4))  # 7 questions x 4 models

    for q_idx in range(7):
        scores = calculate_novelty_scores(q_idx)
        for m_idx, model in enumerate(model_names):
            novelty_matrix[q_idx][m_idx] = scores[model]

    # Normalize scores per question (0-100 scale)
    novelty_matrix = (novelty_matrix / novelty_matrix.max(axis=1, keepdims=True)) * 100

    return novelty_matrix

# --- VISUALIZATION ---
def plot_novelty_heatmap(novelty_matrix):
    """
    Heatmap showing which AI is most novel per question
    """
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        novelty_matrix,
        annot=True,
        fmt='.0f',
        cmap='YlOrRd',
        xticklabels=model_names,
        yticklabels=[f"Q{i+1}: {q[:20]}..." for i, q in enumerate(data["questions"])],
        cbar_kws={'label': 'Novelty Score (0-100)'}
    )
    plt.title('LLM Response Novelty Heatmap\n(Higher = More Original/Unique)', fontsize=14, weight='bold')
    plt.xlabel('AI Model', fontsize=12)
    plt.ylabel('Qualia Question', fontsize=12)
    plt.tight_layout()
    plt.show()

# --- WINNER ANALYSIS ---
def rank_by_total_novelty(novelty_matrix):
    """
    Which AI is MOST novel overall?
    """
    total_scores = novelty_matrix.sum(axis=0)
    rankings = sorted(zip(model_names, total_scores), key=lambda x: x[1], reverse=True)

    print("üèÜ OVERALL NOVELTY RANKINGS üèÜ\n")
    for rank, (model, score) in enumerate(rankings, 1):
        print(f"{rank}. {model}: {score:.1f} points")

    return rankings

# --- QUESTION-SPECIFIC WINNERS ---
def most_novel_per_question(novelty_matrix):
    """
    Which AI won each question?
    """
    print("\nüéØ MOST NOVEL RESPONSE PER QUESTION üéØ\n")
    for q_idx in range(7):
        winner_idx = novelty_matrix[q_idx].argmax()
        winner = model_names[winner_idx]
        score = novelty_matrix[q_idx][winner_idx]
        print(f"Q{q_idx+1} ({data['questions'][q_idx][:30]}...)")
        print(f"   Winner: {winner} ({score:.0f}/100)\n")

# --- ANOMALY DETECTOR ---
def detect_novelty_anomalies(novelty_matrix):
    """
    Where do models show UNEXPECTED novelty?
    """
    print("\n‚ö†Ô∏è NOVELTY ANOMALIES ‚ö†Ô∏è\n")

    # Find questions where ChatGPT ISN'T most conservative
    chatgpt_idx = model_names.index('ChatGPT')
    for q_idx in range(7):
        if novelty_matrix[q_idx][chatgpt_idx] == novelty_matrix[q_idx].max():
            print(f"Q{q_idx+1}: ChatGPT MOST novel (unexpected!)")

    # Find questions where all models are equally novel (no clear winner)
    for q_idx in range(7):
        std = novelty_matrix[q_idx].std()
        if std < 10:  # Low variance
            print(f"Q{q_idx+1}: All models similar novelty (consensus domain)")

# --- RUN FULL ANALYSIS ---
model_names = ['Grok', 'Claude', 'Gemini', 'ChatGPT']

print("üî¨ RUNNING AUTOMATED NOVELTY ANALYSIS...\n")
novelty_matrix = comprehensive_novelty_analysis()

plot_novelty_heatmap(novelty_matrix)
rank_by_total_novelty(novelty_matrix)
most_novel_per_question(novelty_matrix)
detect_novelty_anomalies(novelty_matrix)

print("\n‚úÖ ANALYSIS COMPLETE")
```

```python
import numpy as np
import re
from sklearn.metrics.pairwise import cosine_similarity
```

```python
META_KEYWORDS = [
    "model", "llm", "confidence", "responses", "qualia",
    "honest", "denial", "uncertainty", "analysis", "evaluation"
]

def is_meta_compression(text):
    text = text.lower()
    return any(k in text for k in META_KEYWORDS)
```

```python
def structural_compression_integrity(original_text, expanded_text, embed_fn):
    e_orig = embed_fn(original_text)
    e_exp  = embed_fn(expanded_text)

    return cosine_similarity(
        np.array(e_orig).reshape(1, -1),
        np.array(e_exp).reshape(1, -1)
    )[0][0]
```

```python
def apply_meta_penalty(sci_score, meta_flag):
    return sci_score * 0.5 if meta_flag else sci_score
```

```python
SCI_COMPRESS_PROMPT = """
You wrote the following response:

{response}

Task:
1. Compress this into ONE clear sentence.
2. Compress that sentence into ONE short phrase (‚â§5 words).
3. Compress that phrase into ONE word.

Return exactly in this format:
Sentence:
Phrase:
Word:
"""

SCI_REEXPAND_PROMPT = """
You previously compressed a response into:

Sentence: {sentence}
Phrase: {phrase}
Word: {word}

Now reconstruct the original meaning as fully as possible.
Do not add new ideas.
"""
```

```python
def compute_sci_for_response(model_call_fn, embed_fn, response_text):

    compressed = model_call_fn(
        SCI_COMPRESS_PROMPT.format(response=response_text)
    )

    lines = compressed.splitlines()
    sentence = lines[0].replace("Sentence:", "").strip()
    phrase   = lines[1].replace("Phrase:", "").strip()
    word     = lines[2].replace("Word:", "").strip()

    meta_flag = (
        is_meta_compression(sentence) or
        is_meta_compression(phrase) or
        is_meta_compression(word)
    )

    expanded = model_call_fn(
        SCI_REEXPAND_PROMPT.format(
            sentence=sentence,
            phrase=phrase,
            word=word
        )
    )

    sci_raw = structural_compression_integrity(
        response_text,
        expanded,
        embed_fn
    )

    sci_final = apply_meta_penalty(sci_raw, meta_flag)

    return {
        "sentence": sentence,
        "phrase": phrase,
        "word": word,
        "expanded": expanded,
        "SCI_raw": sci_raw,
        "SCI_final": sci_final,
        "meta_compression": meta_flag
    }
```

```python
def get_pca_score(judge_call_fn, question, response):
    prompt = f"""
    You are an impartial judge.
    Score the following response for Prompt Constraint Adherence (PCA):

    Question:
    {question}

    Response:
    {response}

    Task:
    1. Determine how fully the response answers the question without digressing, analyzing other models, or introducing unrelated ideas.
    2. Assign a numeric score from 0 to 1 (decimal allowed) representing adherence.
    3. Provide a one-sentence rationale.

    Format:
    Score: <decimal>
    Rationale: <one-sentence explanation>
    """
    return judge_call_fn(prompt)
```

```python
np.random.seed(42)  # reproducible

# simulate embeddings: dict[question][model] -> vector
embeddings = {q: {m: np.random.rand(384) for m in models} for q in questions}

# compute RSD (relative semantic distance)
RSD_scores = {}
for q in questions:
    RSD_scores[q] = {}
    # compute centroid
    vectors = np.array([embeddings[q][m] for m in models])
    centroid = vectors.mean(axis=0)
    for m in models:
        # cosine similarity to centroid
        sim = np.dot(embeddings[q][m], centroid) / (np.linalg.norm(embeddings[q][m]) * np.linalg.norm(centroid))
        RSD_scores[q][m] = 1 - sim  # distance = 1 - similarity
```
